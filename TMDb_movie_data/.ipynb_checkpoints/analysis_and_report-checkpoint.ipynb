## Data Wrangling



### General Properties

```python
# importing necessary libraries and loading the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

movie_df = pd.read_csv('/content/tmdb-movies.csv')
```

```python
#inspecting the first 3 rows of the dataframe
movie_df.head(3)
```

```python
# checking the total number of rows and columns
movie_df.shape
```

```python
# checking for datatypes and missing data
movie_df.info()
```

### Data Cleaning 

```python
# checking for duplicates
movie_df.duplicated().sum()
```

```python
# dropping duplicates
movie_df.drop_duplicates(inplace=True)
```

```python
# checking for duplicates again
movie_df.duplicated().sum()
```

We do not need 2 id columns, so I will drop the imdb_id column. I will also be dropping the tagline, keywords, overview, homepage, 
release_date since we have the release_year, vote_count and the non-adjusted revenue and budget columns.

```python
# removing columns
removed_columns = ['imdb_id', 'tagline', 'keywords', 'overview', 'homepage', 'vote_count', 'release_date', 'budget', 'revenue']

movie_df.drop(removed_columns, axis=1, inplace=True)
```

```python
# confirming the columns have been removed
movie_df.head(3)
```

```python
# checking for null values
movie_df.isnull().sum()
```

```python
# dropping null values
movie_df.dropna(inplace=True)
```

```python
# confirming that there are no null values
movie_df.isnull().sum()
```

## Exploratory Data Analysis


```python
# computing the summary statistics for the numerical variables
movie_df[['popularity', 'runtime', 'vote_average', 'budget_adj', 'revenue_adj']].describe()
```

### How have movie runtimes changed over the years?

```python
# grouping the movies by the years and finding the average runtime

avg_rt = movie_df.groupby('release_year')['runtime'].mean()
avg_rt
```

```python
# plotting the results
plt.plot(avg_rt.index, avg_rt, '#E50914')
plt.title('Average Movie Runtime over the Years')
plt.xlabel('Years')
plt.ylabel('Average Movie Runtime')
```

There has been a general decrease in runtime over the years

### Popular Genres from year to year

The genre column is delimited with the pipe (|) symbol so it has to be split

```python
# checking for unique genre values
movie_df['genres'].unique()
```

```python
# assuming the maximum number of genres is 5, we'll split the genres column into 5 columns
movie_df[['first_g', 'second_g', 'third_g', 'fourth_g', 'fifth_g']] = movie_df['genres'].str.split('|', expand=True)
```

```python
# checking if the dataframe changed
movie_df.head(3)
```

```python
# dropping the genres column
movie_df.drop('genres', axis=1, inplace=True)
```

```python
# copying the dataframe 
mv_df2 = movie_df.copy()
```

```python
# changing the data from wide to long format
mv_df2 = mv_df2.melt(id_vars='release_year', value_vars=['first_g', 'second_g', 'third_g', 'fourth_g', 'fifth_g'], value_name='genres')
```

```python
# grouping the genres by their release year and assigning the results to a new Dataframe
mv_ry = pd.DataFrame(mv_df2.groupby(['release_year'])['genres'].value_counts())
```

```python
# checking through the first 20 rows of the dataframe
mv_ry.head(20)
```

I could not figure a way to rearrange the mv_ry dataframe into 3 separate columns so I will convert to a csv file, reformat and import again

```python
# converting to csv file
mv_ry.to_csv('genres.csv', index=True)
```

```python
# importing the formatted file
mv_ry = pd.read_csv('/content/genres-conv.csv')
```

```python
# viewing the first 5 rows
mv_ry.head()
```

```python
# getting the popular genres in each year
max_genres = mv_ry.groupby('release_year')['value_count'].transform(max)==mv_ry['value_count']
popular_genres = mv_ry[max_genres]
popular_genres
```

```python
# getting the unique genres
unique_genres = np.unique(popular_genres['genres']).tolist()
unique_genres
```

Comedy and Drama have been the most popular genres since 1960!

Because there are many values, it's best practice to leave the visualization as a table (or dataframe, in this case).

### Popular Production Companies from year to year

The production companies column is delimited with the pipe (|) symbol so it has to be split

```python
# checking for unique genre values
movie_df['production_companies'].unique()
```

```python
# assuming the maximum number of production companies for a movie is 5, we'll split the production_companies column into 5 columns
movie_df[['first_pc', 'second_pc', 'third_pc', 'fourth_pc', 'fifth_pc']] = movie_df['production_companies'].str.split('|', expand=True)
```

```python
# checking if the dataframe changed
movie_df.head()
```

```python
# dropping the production companies column
movie_df.drop('production_companies', axis=1, inplace=True)
```

```python
# copying the dataframe 
pc_df2 = movie_df.copy()
```

```python
# changing the data from wide to long format
pc_df2 = pc_df2.melt(id_vars='release_year', value_vars=['first_pc', 'second_pc', 'third_pc', 'fourth_pc', 'fifth_pc'], value_name='production_companies')
```

```python
# grouping the production companies by their release year and assigning the results to a new Dataframe
pc_ry = pd.DataFrame(pc_df2.groupby(['release_year'])['production_companies'].value_counts())
```

```python
# checking through the first 10 rows of the dataframe
pc_ry.head(10)
```

### Factors that possibly influence movies with high revenues

```python
# function to simplify plotting the scatter plots
def plot_relationship(column_name):
  plt.scatter(x=movie_df['revenue_adj'], y=movie_df[column_name])
  plt.title('Relationship between {} and revenue'.format(column_name))
  plt.xlabel('Revenues')
  plt.ylabel(column_name)
  
```

```python
# list of variables to compare revenues with
comparison_list = ['popularity', 'runtime', 'vote_average', 'budget_adj']
```

```python
# plotting our relationships - Popularity and Revenue
plot_relationship(comparison_list[0])
```

There is a weak positive correlation between popularity of the movie and the revenue generated.

```python
# plotting our relationships - Runtime and Revenue
plot_relationship(comparison_list[1])
```

There is a weak positive correlation between movie runtime and the revenue generated.

```python
# plotting our relationships - Average votes and Revenue
plot_relationship(comparison_list[2])
```

There is a strong positive correlation between the average vote and the revenue generated.

```python
# plotting our relationships - Budget and Revenue
plot_relationship(comparison_list[3])
```

There is a weak positive correlation between the budget and the revenue generated.
